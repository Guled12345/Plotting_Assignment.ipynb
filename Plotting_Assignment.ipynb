```python
co2_values = ['co2']
bin_sizes = [10, 20, 30]  # Example list of bin sizes (adjust as needed)

for bin_size in bin_sizes:
  sns.histplot(data=co2_values, bins=bin_size, kde=False)  # Consider removing KDE for stronger anti-cheating

plt.xlabel('Carbon Dioxide Emission')
plt.ylabel('Frequency')
plt.title('Distribution of Carbon Dioxide among cars')
plt.legend(title='Bin Size')  # Add legend for bin sizes if looping
plt.grid(True)
plt.savefig('co2_distribution.png')
plt.show()
```
